### Shifting features function

```python
def feature_shifting(features, target):
    
    num_feat=features.shape[1]
    
    lists=[[] for i in range(num_feat)]
    
    feature_lags=[]
    
    # Finding lag values for every feature by using np.corrcoef and adding max lags to feature_lags
    for i in range(num_feat):
        for j in range(1,30):
            lists[i].append((j, abs(np.corrcoef(target[j:],features.iloc[:,i][:-j])[0][1])))
        feature_lags.append(max(lists[i],key=lambda item:item[1])[0])
        fetures_shifted=features.copy()
    
    # shifting every feature according to max lag values
    for i in range(num_feat):
        fetures_shifted.iloc[:,i]=fetures_shifted.iloc[:,i].shift(periods=feature_lags[i])
    
    # cutting NaN parts of features
    lag=max(feature_lags)
    features_final=fetures_shifted.iloc[lag:,:]
    
    target_final=target[lag:]
    df_final=pd.concat([features_final,target_final], axis=1)
    
    return df_final
```

### Shifting features v1.2

```python
def feature_shifting2(features, target, shift_interval_start, shift_interval_end):
    
    num_feat=features.shape[1]
    
    lists=[[] for i in range(num_feat)]
    
    feature_lags=[]
    
    # Finding lag values for every feature by using np.corrcoef and adding max lags to feature_lags
    for i in range(num_feat):
        for j in range(shift_interval_start, shift_interval_end):
            lists[i].append((j, abs(np.corrcoef(target[j:],features.iloc[:,i][:-j])[0][1])))
        feature_lags.append(max(lists[i],key=lambda item:item[1])[0])
        fetures_shifted=features.copy()
    # adding new rows to end of dataframe
    new_tail=pd.DataFrame(columns=fetures_shifted.columns, 
                          index=pd.date_range(fetures_shifted.index[-1]+pd.DateOffset(1), fetures_shifted.index[-1]+pd.DateOffset(7)))
    fetures_shifted=pd.concat([fetures_shifted, new_tail], axis=0)
    
    # shifting every feature according to max lag values
    for i in range(num_feat):
        fetures_shifted.iloc[:,i]=fetures_shifted.iloc[:,i].shift(periods=feature_lags[i])
    
    # cutting NaN parts of features
    lag=max(feature_lags)
    features_final=fetures_shifted.iloc[lag:,:]
    
    target_final=target[lag:]
    df_final=pd.concat([features_final,target_final], axis=1)
    
    return df_final
```
